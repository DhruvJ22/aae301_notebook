# Discrete Fourier Transform

```python
%pylab inline
```

```python
nu = 101
t = linspace(0, 2*pi, nu)
F = fft.fft(np.eye(nu))
print(np.round(F, 1))
```

```python
lam = exp(-1j*2*pi/nu)
z = lam**range(nu)
plt.plot(real(z), imag(z), 'x')
np.round(z, 2)
grid()
axis('equal');
```

```python
lam = exp(-1j*2*pi/nu)
F2 = np.zeros((nu, nu), dtype=complex)
ak_vect = np.hstack([np.arange(0, nu//2 + 1), -np.arange(nu//2, 0, -1)])
j_vect = np.arange(nu)
print('ak:', ak_vect)
print('js:', j_vect)

for k in ak_vect:
    for j in j_vect:
        F2[j, k] = lam**(k*j)
print(np.round(F2, 1))
```

```python
x = 2*sin(10*nu*t) + 2
plt.plot(x)
plt.grid()
```

```python
ak = F.dot(x)
plt.bar(ak_vect, abs(ak))
plt.grid()
plt.xlabel('k')
plt.ylabel('$|a_k|$')
```

```python
ak2 = inv(F2).dot(x)
plt.bar(ak_vect, abs(ak2))
plt.grid()
plt.xlabel('k')
plt.ylabel('$|a_k|$')
```

```python
plot(np.abs(ak - conj(ak2)*nu))
plt.grid()
```
