```python
import sounddevice as sd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fs = 44100
duration = 5
myrecording = sd.rec(int(duration * fs), samplerate=fs, channels=1)
sd.wait()
```

```python
sd.play(myrecording)
sd.wait()
```

```python
t = np.arange(0, 1, 1.0/fs)
data = 0.04*np.sin(1000*2*np.pi*t)
plt.plot(data)
```
